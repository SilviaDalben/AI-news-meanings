This project aims to analyze the overall discourse around the term “Artificial Intelligence” in the past year and evaluate how the press coverage of news outlets in the United States changed with the emergence of generative AI systems such as ChatGPT, Dall-E and Midjourney.

Considering that the press coverage has a high influence over the public opinion, we want to evaluate which terms were associated with Artificial Intelligence over the period analyzed and how the meaning of this term changed over the time.

For data gathering, I used a tool called Mediacloud (https://search.mediacloud.org/) and I run a monthly search of the term “Artificial Intelligence” from July 2022 until June 2023, getting information about the articles referring to this term published by 256 news outlets in the United States.

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

## Analysis 2 | Who published more articles?

First, I will need to join the "content" datasets from each month to one single dataset.

```python
# importing datasets
df_july_2022 = pd.read_csv('july_2022_content.csv')
df_aug_2022 = pd.read_csv('aug_2022_content.csv')
df_sept_2022 = pd.read_csv('sept_2022_content.csv')
df_oct_2022 = pd.read_csv('oct_2022_content.csv')
df_nov_2022 = pd.read_csv('nov_2022_content.csv')
df_dec_2022 = pd.read_csv('dec_2022_content.csv')
df_jan_2023 = pd.read_csv('jan_2023_content.csv')
df_feb_2023 = pd.read_csv('feb_2023_content.csv')
df_march_2023 = pd.read_csv('march_2023_content.csv')
df_april_2023 = pd.read_csv('april_2023_content.csv')
df_may_2023 = pd.read_csv('may_2023_content.csv')
df_june_2023 = pd.read_csv('june_2023_content.csv')
```

### Who is Upworthy?

Upworthy is a storytelling website with a focus on positive aspects, it publishes "good news".

Based in the <b>Silicon Valley</b>, the website was started in March 2012 by Eli Pariser and Peter Koechley with a US$ 4 million funding raised within angel investors that included <b>Facebook</b> co-founder Chris Hughes, <b>BuzzFeed</b> co-founder John Johnson, and <b>Reddit</b> co-founder Alexis Ohanian.

In June 2013, an article in Fast Company called Upworthy "the fastest growing media site of all time". The site popularized a style of two-phrase headlines. It has been criticized for its use of overly sensationalized, emotionally manipulative, "clickbait" style, headlines as well as having a liberal bias, and simplifying issues that are controversial by nature.

Upworthy was accused of laying off one round of writers in 2015, and another in 2016, after an unionization effort by some of the staff. The union involved, the Writers Guild of America, East, has organized several online "viral" news publishers. In January 2017, Upworthy was acquired by Good Worldwide, a company that similarly focused on spreading progressive messages through their websites such as good.is. The newsrooms of the two organizations would merge as part of the acquisition.

It has worked with Unilever, <b>Skype</b>, CoverGirl, and the <b>Bill and Melinda Gates Foundation</b>.

### Therefore, I decided to create a new dataset as well with just the considered more traditional media / institutional press considering the news outlets that published at least 200 articles about Artificial Intelligence during the period analyzed by this study (July 2022 - June 2023).

```python
# convert to datetime
df_content['publish_date'] = pd.to_datetime(df_content['publish_date'], errors='coerce')
df_content['publish_date'] = df_content['publish_date'].dt.date
df_content['publish_date'].head()
```

```python
#filter just the articles published by the traditional press with at least 200 articles published about AI in the period analyzed.

df_traditional_press = df_content[df_content['media_name'].
                                  str.contains("Forbes|Washington Post|Business Insider|CNBC|foxbusiness.com|New York Times|Wall Street Journal|Guardian|Seattle Times|New York Post|techradar.com|CNN|TechCrunch|Gizmodo|ABC News|Fox News|CBS News|CNET|Newsweek|Fortune|LA Times|SF Chronicle|Arkansas Democrat-Gazette")]
df_traditional_press.info()
```

### Here I created a subset (df_traditional_press) with 14,486 (44.56%) articles out of the original dataset that had 32,507 articles.

### From the original list that had 256 news outlets, those articles were published by 25 U.S. news outlets listed in the dataset that I classified as traditional press. The criteria to select them was to have at least 200 articles about Artificial Intelligence published in the period analyzed by this study, which is from July 2022 until June 2023.

<b>They are:</b> Forbes, Business Insider, CNBC, foxbusiness.com, New York Times, Wall Street Journal, Washington Post, Guardian US, Seattle Times, New York Post, techradar.com, CNN U.S. Edition, CNN, Guardian, TechCrunch, Gizmodo, ABC News, Fox News, CBS News, CNET, Newsweek, Fortune, LA Times, San Francisco Chronicle, Arkansas Democrat-Gazette.

```python
df_traditional_press_count = df_traditional_press.groupby(['media_name'])['media_name'].count().sort_values(ascending=False) 
df_traditional_press_count

# Note that Newsweek is duplicated, but this won't really affect our analysis.
```

### Then I want to groupby date to see when these articles were most published...

For that, I am using these links as a resource:

https://www.tutorialspoint.com/how-to-convert-datetime-to-date-in-pandas#:~:text=to_datetime()%20function.,DataFrame%20to%20see%20the%20result.

https://www.geeksforgeeks.org/convert-the-column-type-from-string-to-datetime-format-in-pandas-dataframe/

```python
df_tradpress_date = df_traditional_press.groupby(['publish_date'])['publish_date'].count().sort_values(ascending=False) 
df_tradpress_date.head(10)
```

```python
date = ["June 22", "May 25", "May 18", "May 30", "May 16",
        "May 10", "June 1", "April 25","May 31", "June 14"]

articles_published = [149, 148, 143, 131, 131, 124, 122, 121, 121, 121]

# Create a bar chart
plt.figure(figsize=(10, 7))
bars = plt.bar(date, articles_published, color='lightskyblue')

plt.xlabel('', fontsize = 14)
plt.ylabel('Number of Articles Published', fontsize = 14)
plt.title('Days with most articles published by U.S. traditional media outlets')

for bar, articles_published in zip(bars, articles_published):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.01, bar.get_height() + 1, str(articles_published), ha='center', color='black')


title_string = "Days with more AI articles published"
subtitle_string = "Days with most articles published by U.S. traditional media outlets with at \nleast 200 articles published within the period analyzed. (July 2022 - June 2023)"
plt.suptitle(title_string, y=1, fontsize=20)
plt.title(subtitle_string, y=1, fontsize=13)
plt.annotate('Source: Mediacloud | by Silvia DalBen Furtado',
            xy = (1.0, -0.3),
            xycoords='axes fraction',
            ha='right',
            va="center",
            fontsize=8)

plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### First you see in this analysis that you don't have an outlier, a date within the period analyzed in which more articles were published. it is quite stable, as we are looking for the top 10 days in one year with 365 days.

### Second, it is interesting to notice that all articles were published between April 25, 2023 and June 22, 2023 (top1), the last three months of the 12 month dataset analyzed. So this corroborates with the first plot that shows that the number of articles published is increasing over the time.
